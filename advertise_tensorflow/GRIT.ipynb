```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers # type: ignore
from tensorflow.keras.models import Model # type: ignore

# 模拟数据：用户ID、广告ID和点击情况
user_ids = np.array([1, 1, 1, 1, 1, 1, 1, 1, 1])
ad_ids = np.array([1, 2, 2, 1, 1, 1, 1, 2, 2])
clicks = np.array([1, 0, 0, 1, 1, 1, 1, 0, 0])  # 1表示点击，0表示没有点击

# 定义输入层：用户ID和广告ID
#layers.Input(shape=None, batch_size=None, name=None, dtype=None, sparse稀疏张量=False, tensor预先存在的张量=None, ragged不规则张量=False, **kwargs)
user_input = layers.Input(shape=(1,), dtype=tf.int32, name="user_id")
ad_input = layers.Input(shape=(1,), dtype=tf.int32, name="ad_id")

# 创建嵌入层（Embedding Layer）：为用户和广告ID创建低维向量表示
#layers.Embedding(input_dim, output_dim, embeddings_initializer='uniform', embeddings_regularizer=None, activity_regularizer=None, embeddings_constraint=None, mask_zero=False, input_length=None)
user_embedding = layers.Embedding(input_dim=10, output_dim=2)(user_input)  # 6是用户的总数，4是嵌入维度
ad_embedding = layers.Embedding(input_dim=10, output_dim=2)(ad_input)  # 6是广告的总数，4是嵌入维度

# 扁平化嵌入层的输出，将卷积层或嵌入层的输出转换为一维向量，以便传递给全连接层
#layers.Flatten(data_format=None)
user_vec = layers.Flatten()(user_embedding)
ad_vec = layers.Flatten()(ad_embedding)

# 合并用户向量和广告向量
merged = layers.concatenate([user_vec, ad_vec])

# 添加全连接层
#layers.Dense(units输出空间维度, activation激活函数=relu/sigmoid, use_bias偏置项=True, kernel_initializer='glorot_uniform', bias_initializer='zeros', kernel_regularizer=None, bias_regularizer=None, activity_regularizer=None, kernel_constraint=None, bias_constraint=None)
x = layers.Dense(8, activation='relu')(merged)
output = layers.Dense(1, activation='sigmoid')(x)  # 使用sigmoid激活函数来预测点击概率

# 构建和编译模型
#Model(inputs=None, outputs=None, name=None)
#model.compile(optimizer优化器='adam'/'sgd', loss损失函数='binary_crossentropy', metrics评估指标=['accuracy']，loss_weights=None, sample_weight_mode=None, weighted_metrics=None, target_tensors=None, distribute=None, **kwargs)
model = Model(inputs=[user_input, ad_input], outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 模型概况,打印模型的结构和参数数量
model.summary()

# 训练模型
#model.fit(x输入数据, y目标数据, batch_size每个批次样本数默认32=None, epochs训练轮数=1, verbose日志显示模式=1, callbacks=None, validation_split验证数据比例=0.0, validation_data=None, shuffle打乱数据=True, class_weight=None, sample_weight=None, initial_epoch=0, steps_per_epoch=None, validation_steps=None, validation_batch_size=None, validation_freq=1, max_queue_size=10, workers=1, use_multiprocessing=False)
model.fit([user_ids, ad_ids], clicks, epochs=20, batch_size=1, verbose=1)

# 模拟一个新的用户和广告对，预测是否点击,
#使用训练好的模型进行预测，model.predict(x输入数据, batch_size每个批次样本数默认32=None, verbose日志显示模式=0, steps=None, callbacks=None, max_queue_size=10, workers=1, use_multiprocessing=False)
new_user_id = np.array([1])
new_ad_id = np.array([1])
prediction = model.predict([new_user_id, new_ad_id])

print(f"预测点击概率：{prediction[0][0]:.4f}")
```
